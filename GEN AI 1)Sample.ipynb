```python
import nltk
from nltk.tag import hmm
from nltk.probability import LidstoneProbDist
```

```python
# Training data
train_data = [
    [('walk', 'Sunny'), ('shop', 'Sunny'), ('clean', 'Rainy')],
    [('walk', 'Sunny'), ('walk', 'Sunny'), ('shop', 'Rainy')],
    [('clean', 'Rainy'), ('shop', 'Rainy'), ('walk', 'Sunny')],
    [('walk', 'Sunny'), ('clean', 'Rainy'), ('clean', 'Rainy')],
    [('shop', 'Sunny'), ('walk', 'Sunny'), ('clean', 'Rainy')],
    [('clean', 'Rainy'), ('clean', 'Rainy'), ('shop', 'Rainy')]
]
```

```python
# Train HMM model
trainer = hmm.HiddenMarkovModelTrainer()
model = trainer.train_supervised(
    train_data,
    estimator=lambda fd, bins: LidstoneProbDist(fd, 0.1, bins)
)
```

```python
# Test sequence
test_seq = ['walk', 'shop', 'clean']

# Viterbi prediction
predicted_states = model.tag(test_seq)

# Sequence probability (Forward algorithm)
prob = model.probability(test_seq)
```

```python
# Output
print("Predicted states:", predicted_states)
print("Sequence Probability:", prob)
print(model)
```
